2. Why do raw strings often appear in Regex objects?

 - Raw strings are used so that backslashes do not have to be escaped
 - In contrast to a conventional string, a raw string considers backslashes () as literal characters. When working with strings that include a lot of backslashes, such as regular expressions or directory paths on Windows, raw strings are helpful.
